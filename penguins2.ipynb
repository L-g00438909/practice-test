```python
# Assuming df is your DataFrame and 'species' is the name of the column containing species information
# 'species_value' is the specific species you want to filter for

# Filter rows for a specific species
species_data = df[df['species'] == 'Adelie']

# Print the data for the specific species
print(species_data)
```

```python
# Assuming df is your DataFrame and 'species' is the name of the column containing species information
# 'species_value' is the specific species you want to filter for

# Filter rows for a specific species
species_data = df[df['species'] == 'Gentoo']

# Print the data for the specific species
print(species_data)
```

```python
# Assuming df is your DataFrame and 'species' is the name of the column containing species information
# 'species_value' is the specific species you want to filter for

# Filter rows for a specific species
species_data = df[df['species'] == 'Chinstrap']

# Print the data for the specific species
print(species_data)
```

```python
# Assuming df is your DataFrame and 'species' is the name of the column containing species information
# Use value_counts() to count the occurrences of each species
species_counts = df['species'].value_counts()

# Print the counts of each species
print(species_counts)
```

```python
import matplotlib.pyplot as plt

# Plot the counts of each species
species_counts.plot(kind='bar')
plt.title('Counts of Each Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```
